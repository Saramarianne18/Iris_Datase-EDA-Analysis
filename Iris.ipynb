### 1. Introduction

- The objective of this analysis is to explore the Iris dataset to identify patterns, relationships, and key features that distinguish different species. The analysis focuses on understanding data distribution, correlations, and feature importance to support data-driven insights.

### 2. Tools & Dataset
- Tools: Python (Pandas, Matplotlib, Seaborn)
- Dataset: Iris Dataset

### 3. Loading and Prearing Dataset

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Convert labels
df['species'] = df['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})

df.head()
```

### 4. Data Overview

```python
df.info()
df.describe()
```

##### Insight:
- Dataset contains 150 rows and 5 columns
- Features are numerical and suitable for analysis

### 5. Summary Statistics 

```python
df.groupby('species').mean()
```

##### Insight:
- Setosa has smaller petal measurements
- Virginica has the largest measurements

```python
# Summary Statistics
df.info()
print('\n')
print(df.describe())
```

```python
# Species Distribution
print(df['species'].value_counts())
```

```python
# Correlation Heatmap
corr = df.drop('species', axis=1).corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(10,6))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='species', y=col, data=df)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Histograms
plt.figure(figsize=(10,6))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

# EDA Report

## Summary
The Iris dataset consists of 150 samples of iris flowers from three species: Setosa, Versicolor, and Virginica. Each sample has four features: sepal length, sepal width, petal length, and petal width.

## Key Findings
- **Dataset Balance**: The dataset is perfectly balanced with 50 samples for each species.
- **Correlations**: 
  - Petal length and petal width show the highest correlation (0.96).
  - Sepal length correlates moderately with petal length (0.87) and petal width (0.82).
  - Sepal width has low correlation with other features.
- **Species Patterns**:
  - Setosa species has distinctly smaller petal dimensions compared to Versicolor and Virginica.
  - Versicolor and Virginica show some overlap in sepal measurements but are more separable by petal features.
  - Setosa can be easily distinguished from the other two species based on petal length (< 2 cm) and petal width (< 0.8 cm).
- **Distributions**: Most features show roughly normal distributions, with some skewness in petal measurements.

## Visual Insights
- The pairplot reveals clear clusters for Setosa and partial separation for the other species.
- Boxplots confirm Setosa's distinct characteristics and highlight outliers in Versicolor and Virginica.
- Histograms show the spread of measurements across species.

This EDA provides a foundation for further analysis, such as classification modeling.
